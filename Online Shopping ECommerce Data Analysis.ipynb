# Online shopping/E-Commerce Data Analysis 

```python
import numpy as np
import os
os.getcwd()
```

```python
com_data = np.genfromtxt("DS1_C4_S1_Shopping_Data_Challenge.csv",delimiter=",",skip_header = 1, dtype="int")
com_data
```

### 

#### TASK-1 : Display the details of the customers whose spending score is greater then 80. also, display the total count of these customers.

```python
index = np.where(com_data[:,3] > 80)
Cust = com_data[:,:][index]
print("Customer details whose soending score is greater than 80: \n",Cust)
print()
print("Count of customers: ",Cust[:,1].size)
```

### 

#### TASK-2 : Extract and display the data of the customers who are 20 to 24 years of age.

```python
index1 = np.where((com_data[:,1] >=20) & (com_data[:,1] <=24))
age_data = com_data[:,:][index1]
print("Customers who have 20 to 24 years of age: \n",age_data)
```

### 

#### TASK-3 : Display the IDs of the customers with the lowest annual income(s) and the highest annual income.

```python
lowest = np.min(com_data[:,2])
index2 = np.where(com_data[:,2]==[lowest])
lowest_income = com_data[:,:][index2]
highest = np.max(com_data[:,2])
index3 = np.where(com_data[:,2]==[highest])
highest_income = com_data[:,:][index3]
print("Customers with Lowest annual Income: \n",lowest_income)
print()
print("Customers with Highest annual Income: \n",highest_income)
```

### 

#### TASK-4 : Display the details of the  customers who are less than 45 years of age, have an annual income greater than 35K, and have a spending score greater than 75. how many such customers are there in the given data?

```python
index4 = np.where((com_data[:,1] < 45) & (com_data[:,2] >35) & (com_data[:,3] >75))
customer = com_data[:,:][index4]
print("Details of Customers: \n",customer)
print()
print("Number of students: ",customer[:,1].size)
```
