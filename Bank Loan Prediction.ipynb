### TASK - 7 : Calculate and Display average the income of customers based on each area where their property is located(urban, semiurban, rural).

```python
urban_data = bank_data[bank_data["Property_Area"]=="Urban"]
avg_urban_data = urban_data["ApplicantIncome"].mean()
print("Average income of customers from Urban area : ",avg_urban_data)
```

```python
semiurban_data = bank_data[bank_data["Property_Area"]=="Semiurban"]
avg_semiurban_data = semiurban_data["ApplicantIncome"].mean()
print("Average income of customers from Semiurban area : ",avg_semiurban_data)
```

```python
rural_data = bank_data[bank_data["Property_Area"]=="Rural"]
avg_rural_data = rural_data["ApplicantIncome"].mean()
print("Average income of customers from Rural area : ",avg_rural_data)
```
