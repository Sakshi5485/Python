```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
emp_data = pd.read_excel("DS1_C4_S5_Employee_Data_Practice.xlsx")
```

### 

## TASK - 1 :

### What is the total annual salary for the all the employees of each department ? Display the results using bar graph.

```python
total_sal = emp_data.groupby(by="Department")["Annual Salary ($)"].sum()
total_sal
```

```python
total_sal.plot(x = 'Annual Salary ($)',
              kind = 'bar',
              stacked = False,
              title = 'Department-wise annual salary',xlabel = 'Annual Salaries',ylabel = 'Departments')
```

#### Interpretation : IT department has Highest annual salary and Finance department has lowest annual salary.

# TASK - 2 :

### The company wants to analyze gender equality across several departments. Display the department-wise total number of male and female employees using DataFrame.

```python
gender_count = emp_data.groupby(by=["Department","Gender"])["Employee_Code "].count()
Department_Name = pd.DataFrame(gender_count).unstack(level=1)
Department_Name
```

```python
dept_data = {
    'Departments':['Finance','HR','IT','Sales'],
    'Female_count' : [2,11,10,4],
    'Male_Count' : [11,4,12,16]
    }
dept_data = pd.DataFrame(dept_data)
```

```python
dept_data.plot(x='Departments',
               kind='bar', 
               stacked=True,
               title = 'Gender Equality across several Departments',
               xlabel='Department',ylabel=['Gender','Employee_Code '])
```

#### Interpretation : The IT department has nearly equal number of male and female employees.
