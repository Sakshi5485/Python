## TASK - 2 :

### Thomas, a potential customer from california, is interested in buying a low-priced car, manufactured in 2014. John needs to identify 100 lowest priced cars from the year 2014 with CA as the state of registration. Help John create and display the details of potential cars.

```python
thomas = car_data[(car_data["year"]== 2014) & (car_data["state"]== "ca")]
Thomas =  thomas.sort_values(["mp"]).head(100)
Thomas
```

## 

## TASK - 3 :

### Thomas is confused looking of 100 cars. Help John to display the output in a cross table based on the car manufactures and prices.

```python
pd.crosstab(Thomas.make, Thomas.mp, rownames = ["make"], colnames = ["mp"])
```

## 

## TASK - 4 :

### Thomas is particularly interested in Chevrolet cars. Eric is willing to offer a 10% discount on Chevrolet cars from the list of 100 cars identified in the previous tasks. Display the updated with only chevrolet cars with their original prices and discounted prices.

```python
CH_cars = Thomas[Thomas["make"]=="Chevrolet"]
discount = CH_cars["mp"]*0.9
CH_cars = pd.concat([CH_cars["mp"],discount],axis=1)
CH_cars.columns = ["Original_Price","Discounted_Price"]
CH_cars
```

## 

## TASK - 5 :

### Thomas and several other customers seem to be interested in the cars with black or gray colored interiors. John wants to figure out the cars with interiors in black or gray are priced higher or lower than cars having interiors of other colors. Determine the average price of all the 100 lowest priced cars from the year 2014 based on their interior colors.

```python
black = Thomas[Thomas["interior"]=="black"]
gray = Thomas[Thomas["interior"]=="gray"]
others = Thomas[(Thomas["interior"]!="black") & (Thomas["interior"]!="gray")]
print("Count of cars with black color interior : ",black["year"].size)
print("Count of cars with gray color interior : ",gray["year"].size)
print("Count of cars with other colors interior : ",others["year"].size)
```

```python
black_avg = black["mp"].mean()
gray_avg = gray["mp"].mean()
others_avg = others["mp"].mean()
print("Average price of cars with black color interior :",black_avg)
print("Average price of cars with gray color interior :",gray_avg)
print("Average price of cars with other colors interior :",others_avg)
```

#### The other color interior cars are priced high than black & gray colors interior cars based on their average price.

## TASK - 6 :

### Most buyers to prefer white and black colored cars. John wants o analyze color-wise availability of cars, for each manufacturer from the year 2010-15.

```python
A = ["black","white"]
John = car_data[(car_data["year"]>=2010) & (car_data["year"]<=2015) & (car_data["color"].isin(A))]
John = John.groupby(by=["make","color"])["color"].count()
John = pd.DataFrame(John)
John.columns=["count"]
John
```

## 

## TASK - 7 : 

### Eric, another customer is interested in a good condition BMW Sedan manufactured between 2010 and 2015. John wants to display the details of all such cars that have a condition score >=4.5. Help John to figure out of the year-wise maximum expected price(mp) of the above BMW cars manufactured between the year 2010 and 2015 grouped by each model.

```python
BMW = car_data[(car_data["make"]=="BMW") & (car_data["body"]=="Sedan") & (car_data["year"]>=2010) &
               (car_data["year"]<=2015) & (car_data["condition"]>=4.5)]
BMW_max = BMW.groupby(by=["model","year"])["mp"].max()
BMW_max
```

## TASK - 8 :

### John wants to recommend Eric to also consider SUVs from other manufacturers before finalizing the car. John needs to list the top 100 SUVs available between the years 2010 and 2015, based on their prices(in descending order) with an odometer reading <=30000.

```python
SUV = car_data[(car_data["body"]=="SUV") & (car_data["year"]>=2010) & (car_data["year"]<=2015) & (car_data["odometer"]<=30000)]
SUV = SUV.sort_values(by=["mp"],ascending=False).head(100)
SUV
```
