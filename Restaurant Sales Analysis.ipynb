```python
import numpy as np
import pandas as pd
```

#### TASK-1 : Sort the data in ascending order of sales and display the results

```python
data_asc = data.sort_values(by="Sales")
data_asc
```

### 

#### TASK - 2 : Display the all rows where the sale amount is >=45000

```python
a = data[data["Sales"]>=45000]
a
```

### 

#### TASK - 3 : Filter the data separately for the state of California, Texas and Montana and save them in three different data frames. sort the three data frames in increasing order of sale amount and display them separately.

```python
california_data = data[data["State"]=="California"]
california_data.sort_values("Sales")
```

```python
texas_data = data[data["State"]=="Texas"]
texas_data.sort_values("Sales")
```

```python
montana_data = data[data["State"]=="Montana"]
montana_data.sort_values("Sales")
```

## 

#### TASK - 4 : Calculate the sum of sales (of all six months) for the state specified in Task3 and display the state-wise total sales.

```python
sum_california = sum(california_data["Sales"])
print("Total sales in State of California : ",sum_california)
sum_texas = sum(texas_data["Sales"])
print("Total sales in State of Texas : ",sum_texas)
sum_montana = sum(montana_data["Sales"])
print("Total sales in State of Montana : ",sum_montana)
```

### 

#### TASK - 5 : Display the months with the highest and the lowest monthly sale amount, for the store in the state of California.

```python
min_sale = california_data["Sales"].min()
```

```python
max_sale = california_data["Sales"].max()
california_data[california_data["Sales"] == max_sale]
```

#### TASK - 6 : Create and display a new data frame consisting of state-wise total sales.

```python
Alaska_data = data[data["State"]=="Alaska"]
```

```python
New_Mexico_data = data[data["State"]=="New Mexico"]
```

```python
dic = {"State":["Alaska","Texas","California","Montana","New Mexico"],"Sale in thousand":[sum(Alaska_data["Sales"]),
                                                                                    sum(texas_data["Sales"]),
                                                                                   sum(california_data["Sales"]),
                                                                                   sum(montana_data["Sales"]),
                                                                                   sum(New_Mexico_data["Sales"])]}
Total_sales = pd.DataFrame(dic)
Total_sales
```

#### TASK-7 : Sort the rows in the data frame created in Task 6 by the sales amount. save the data frame as a total_sales.csv file in your local system. check your current working directory to see if the file is saved correctly.

```python
total_sale = Total_sales.sort_values(by="Sale in thousand")
total_sale
```

```python
total_sale.to_csv("total_sales.csv")
```
